# Fault predictions from sensor data

## About
### Goal: 
Avoid product defects in finished product for global electrical component manufacturer

### Method:
- Early detection of unfinished products
- Data from early stages in production process

### Details:
- Pipeline with multiple stage
- Sensor readings for first 4 stages
- Predict QA at end of pipeline for finished process

### LUSY dataset:
- Primary: Random checks in different stages
- Secondary: Bi-secondly sensor readings

_note: join on produdct ID, but only use secondary data from earlier points in time_
