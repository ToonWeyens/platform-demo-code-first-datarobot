## Set up and configure connection to SDK

```python
import datarobot as dr #needs 2.25 or later
import pandas as pd
import pickle
```

```python
dr.Client(config_path = '/Volumes/GoogleDrive/My Drive/Configurations/drconfig.yaml')
```

## Set up datasets in AI Catalog
You can use a variety of databases:



Here we will use an S3 bucket.

```python
dataPath = 'https://demo-datarobot.s3.eu-central-1.amazonaws.com/'
```

```python
primary_dataset = dr.Dataset.create_from_url(url = dataPath+'AFD_qc_1_b_ts.csv')
```

```python
secondary_dataset = dr.Dataset.create_from_url(url = dataPath+'AFD_qc_2_ts.csv')
```

## Save datesets

```python
datasets = dict()
datasets['primary'] = primary_dataset
datasets['secondary'] = secondary_dataset
```

```python
filename = 'datasets.pickle'
outfile = open(filename,'wb')
pickle.dump(datasets,outfile)
outfile.close()
```
